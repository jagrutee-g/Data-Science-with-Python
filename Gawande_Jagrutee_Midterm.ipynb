###### Question 3: Comapring two stocks Apple and Google

1. A normal describe will help determine the min max of a stock.
2. A moving average over the years will also help determine the overall stock price
3. A graph plot of the same will help understand how the stock prices have performed over the years
4. Also the daily returns of Google will help us understand which months to focus on.

```python
aapl.describe()
```

```python
googl.describe()
```

```python
aapl.set_index('Date', inplace=True)
googl.set_index('Date', inplace=True)
stocks2 = pd.DataFrame({"AAPL": aapl["Close"],
                      "GOOG": googl["Close"]})

ax = stocks2.plot(grid = True,figsize=(10,10), title="Stock Price Comparison of Apple and Google")
ax.set_ylabel("Stock Price")
```

We can see that the Google stocks did better that Apple over the years

We will explore Google more in detail

```python
googl2 = googl[googl['Date']>='2017-09-01']
date_range = googl2['Date']

fig = plt.figure()
fig.set_size_inches(18.5, 10.5)
ax=fig.add_subplot(111,title="Moving Average Plot")

ax.plot(date_range,googl2['Close'].values,label="Original")
ax.plot(date_range,pd.rolling_mean(googl2,20)["Close"].values)
fig.autofmt_xdate(rotation=90)
```

From the above plot we can see that the on average the Google stocks are steadily moving up for the said time range.

```python
googl2 = googl[googl['Date']>='2017-01-01'].copy() 
googl2.set_index('Date',inplace=True)

# Using pct_change to find the percent change for each day
googl2['Daily Return'] = googl2['Close'].pct_change()

# Plotting the daily return percentage
ax = googl2['Daily Return'].plot(figsize=(18,10), legend=True, linestyle='--', marker='o')
```
